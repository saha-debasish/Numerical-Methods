```python
def factorial(n):
    """
    This is a function that gives the factorial
    of any positive integer. The formula is
    N! = N*(n-1)*(N-2)* ... *3*2*1
    """
    fact = 1
    for i in range(1,n+1):
        fact = fact*i
        
    return fact
```

```python
def sin(x):
    """
    sin(x) = x - x^3/3! + x^5/5! - ... + (-1)^k*x^(2k+1)/(2k+1)!
    """
    N = 20
    sine_sum = 0
    for i in range(N):
        sine = ((-1)**i)*x**((2*i)+1)/(factorial((2*i)+1))
        sine_sum = sine_sum + sine
    return sine_sum
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0.0,4.0*np.pi,100)
sin = sin(X)

plt.plot(X,sin,label="sin(x)")
plt.legend()
plt.show()
```

```python
def cos(x):
    """
    cos(x) = 1 - x^2/2! + x^4/4! - ... + (-1)^k*x^(2k)/(2k)!
    """
    N = 20
    cos_sum = 0
    for i in range(N):
        cos = ((-1)**i)*x**(2*i)/(factorial(2*i))
        cos_sum = cos_sum + cos
    return cos_sum
```

```python
X = np.linspace(0.0,4.0*np.pi,100)
cos = cos(X)

plt.plot(X,cos)
plt.legend(["cos(x)"])
plt.show()
```

```python
plt.plot(X,sin,color="r",label="sin(x)")
plt.plot(X,cos,color="b",label="cos(x)")
plt.legend()
plt.xlabel("x")
plt.ylabel("sin(x), cos(x)")
plt.show()
```
